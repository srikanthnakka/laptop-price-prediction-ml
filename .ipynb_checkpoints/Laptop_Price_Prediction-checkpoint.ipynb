```python
import re # importing regular expression

# Function to extract resolution width and height
def extract_resolution(resolution):
    match = re.search(r'(\d+)x(\d+)', resolution)
    if match:
        return int(match.group(1)), int(match.group(2))
    else:
        return None, None

# Apply the function to the 'ScreenResolution' column to separate the screen resolution 
laptop_data['Resolution_Width'], laptop_data['Resolution_Height'] = zip(*laptop_data['ScreenResolution'].apply(extract_resolution))
```

```python
# Add new columns for Memory categories directly to the DataFrame
import re

def parse_storage_size(storage_str):
    """
    Parse storage string and extract size in GB for specific storage type
    """
    if pd.isna(storage_str) or storage_str == '?':
        return 0
        
    def convert_to_gb(size_str):
        if 'TB' in size_str:
            return float(size_str.replace('TB', '').strip()) * 1024
        return float(size_str.replace('GB', '').strip())
    
    size_match = re.search(r'(\d+(?:\.\d+)?(?:GB|TB))', storage_str)
    if size_match:
        return convert_to_gb(size_match.group(1))
    return 0

# Create new columns initialized with 0
laptop_data['SSD'] = 0
laptop_data['HDD'] = 0
laptop_data['Flash_Storage'] = 0
laptop_data['Hybrid'] = 0

# Process each row
for idx, row in laptop_data.iterrows():
    storage_devices = str(row['Memory']).split(' + ')
    
    for device in storage_devices:
        size = parse_storage_size(device)
        
        if 'SSD' in device:
            laptop_data.at[idx, 'SSD'] = size
        elif 'HDD' in device:
            laptop_data.at[idx, 'HDD'] = size
        elif 'Flash Storage' in device:
            laptop_data.at[idx, 'Flash_Storage'] = size
        elif 'Hybrid' in device:
            laptop_data.at[idx, 'Hybrid'] = size

# memory have different types and different storage methods to simply that 
#add new columns ssd,Hdd,flash_storage,Hybrid to divide and see the price variations
```

## Model Development

```python
#separate the feature variables and label variables 
x=laptop_data.drop(columns=['Price'])#selecting the all feature variables expecting the label value
y=np.log(laptop_data['Price'])#select the label data
```

```python
x #Feature varibles
```

* Raw data cannot be directly used for machine learning models.  
* In this section, the dataset is transformed into a suitable format by handling categorical variables, scaling numerical features, and preparing the data for model training.

    #Categorical Feature Encoding

* Machine learning models require numerical inputs.  
* Categorical features such as brand, processor type, operating system, and storage type were encoded into numerical values using appropriate encoding techniques.
* This transformation allows the model to learn patterns associated with different categories.


```python

# One-hot encode categorical variables
X = pd.get_dummies(
    x,
    columns=['Company', 'TypeName', 'Cpu_brand', 'Gpu_brand', 'OS'],  # Specify columns to encode
    drop_first=True  # Avoid dummy variable trap
)

X.head()
#using get dummies to encoding the categorical variables
```

### Train-Test Split

* The dataset was split into training and testing sets to evaluate model performance on unseen data.  
* This helps assess the generalization ability of the model and prevents overfitting.


```python
#Split the test data and train data using sklearn library
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

### Feature Scaling

* Numerical features were scaled to bring them to a similar range.  
* Feature scaling helps improve model convergence and ensures that no single feature dominates due to its scale.


```python
#Feature scaling to avoid misdata lekage and misinterception
from sklearn.preprocessing import StandardScaler
# Define the numerical columns for scaling
numerical_columns = ['Ram', 'Weight', 'Display', 'SSD', 'HDD']

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on X_train and transform it
X_train = X_train.copy()  # Copy to avoid modifying the original data
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])

# Print the scaled training data
X_test[numerical_columns] = scaler.transform(X_test[numerical_columns])
```

- Categorical variables were encoded into numerical form.
- Numerical features were scaled where necessary.
- The dataset was split into training and testing sets.
- The processed data is now suitable for regression modeling.


```python
#import the sklearn modules to evaluate the model using mean absolute error and mean squared error and R2 score 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression #import the linear regression model to develop the model
```

### 🔹 Gradient Boosting Regressor

Gradient Boosting is an ensemble learning technique that builds models sequentially, where each new model corrects the errors made by the previous ones.

It is particularly effective for regression problems involving complex, non-linear relationships and feature interactions, making it well-suited for laptop price prediction.


```python
#importing Gradient Booster regressor to run the model to predict the laptop price
from sklearn.ensemble import GradientBoostingRegressor
#select default parameters to run the model
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Train the model
gb_regressor.fit(X_train, y_train)

# Make predictions
y_pred = gb_regressor.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

# Output the results
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```

* The Gradient Boosting model achieved better performance compared to baseline models such as Linear Regression.  
* Its ability to capture non-linear patterns and reduce bias through iterative learning contributed to improved prediction accuracy.
* This model also low cost function and good accuaracy with around 85%
* Next to get the good parameters we use hyper parameter prune to get best parameters to get good model

### Hyperparameter Tuning – Gradient Boosting

Hyperparameter tuning was applied to the Gradient Boosting model to optimize learning rate, number of estimators, and tree depth.

This improves predictive accuracy while controlling overfitting.


```python
#Hyperparameter Tuning: Fine-tune the selected model to achieve optimal performance.
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50,100,150,200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=GradientBoostingRegressor(), param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and model
best_model = grid_search.best_estimator_

# Make predictions with the best model
y_pred_best = best_model.predict(X_test)

# Calculate evaluation metrics
mae_best = mean_absolute_error(y_test, y_pred_best)
mse_best = mean_squared_error(y_test, y_pred_best)
rmse_best = mse_best**0.5
r2_best = r2_score(y_test, y_pred_best)

# Output the best parameters and performance
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"MAE: {mae_best}")
print(f"MSE: {mse_best}")
print(f"RMSE: {rmse_best}")
print(f"R2 Score: {r2_best}")
```

### 🔹 Hyperparameter Tuning – Random Forest

Hyperparameter tuning was performed to optimize the Random Forest model by adjusting parameters such as the number of trees, maximum depth, and minimum samples per split.

This helps improve model performance and generalization on unseen data.


```python
#Hyper parameter tuning with GridsearchCv
rf = RandomForestRegressor(random_state=42)

# Set up the hyperparameter grid
param_grid = {
    'n_estimators': [50,100,150,200],   # Number of trees in the forest
    'max_depth': [10, 15,20,25, 30, None],   # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],   # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],     # Minimum samples required to be at a leaf node
    'bootstrap': [True, False]         # Whether to use bootstrap sampling
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best model
best_rf = grid_search.best_estimator_

# Make predictions with the best model
y_pre = best_rf.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pre)
mse = mean_squared_error(y_test, y_pre)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pre)

# Print the best hyperparameters and evaluation metrics
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```


### Final Model Selection

After comparing all models, the tuned Gradient Boosting Regressor achieved the best performance based on evaluation metrics such as R², MAE, and RMSE.

Therefore, it was selected as the final model for predicting laptop prices.
 

```python

from sklearn.ensemble import GradientBoostingRegressor

gb_regressor = GradientBoostingRegressor(n_estimators=200,learning_rate=0.1,max_depth= 5,min_samples_leaf= 2,min_samples_split= 5,random_state=42)

# Train the model
gb_regressor.fit(X_train, y_train)

# Make predictions
y_pred = gb_regressor.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

# Output the results
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```
